```python
 import scipy.stats as stats


#H0. 	Deleting less related values
#H1.	XGboosting with all values

#XGboosting Deleting less related values - MSE
result1 = [868.955131,801.173361,875.949622]
# XGBoosting all values - MSE
result2 = [580.327094,742.634755,621.939617]

results = stats.wilcoxon(result1, result2)
print(results)

#WilcoxonResult(statistic=0.0, pvalue=0.25)

# Sí pvalue=0.25 > alpha=0.05, entonces se acepta H0

'''Dado que el valor p es 0.25, que es mayor que el nivel de significancia α de 0.05, no hay suficiente evidencia para rechazar la hipótesis nula (H0). Esto significa que, a eliminación de valores menos relacionados no mejora significativamente el rendimiento del modelo en términos de MSE en comparación con el uso de todos los valores. Por lo tanto, la estrategia de eliminar menos relacionados no muestra un beneficio claro en la reducción del error cuadratico medio, según los resultados del test de Wilcoxon.'''
```
